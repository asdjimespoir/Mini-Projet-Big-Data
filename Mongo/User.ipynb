```python
#get user and role
use Ehealth
db.getUser("Chercheur")
db.getRole( "readWrite", { showPrivileges: true } )

#Revoke and attribute roles
db.revokeRolesFromUser(
    "Chercheur",
    [
      { role: "read", db: "Ehealth" }
    ]
)

db.grantRolesToUser(
    "Chercheur",
    [
      { role: "read", db: "Ehealth" }
    ]
)

#Drop role et user
db.dropUser("Chercheur", {w: "majority", wtimeout: 5000})
db.dropRole( "roleChercheur", { w: "majority" } )
```

```python
#Create role on Ehealth
use Ehealth
db.createRole(
 {
  role: "roleMedecin",
  privileges: [
   {resource: {db: "Ehealth", collection: "Normal"}, actions: [ "find", "update", "insert", "remove" ] },
   {resource: {db: "Ehealth", collection: "Alerte"}, actions: [ "find", "update", "insert", "remove" ] }
  ],
  roles: [{ role: "readWrite", db: "Ehealth"}]
 }
);

use Ehealth
db.createRole(
 {
  role: "roleInfirmier",
  privileges: [
   {resource: {db: "Ehealth", collection: "Normal"}, actions: [ "find", "update", "insert" ] },
   {resource: {db: "Ehealth", collection: "Alerte"}, actions: [ "find", "update", "insert" ] }
  ],
  roles: [{ role: "read", db: "Ehealth"}]
 }
);


db.createRole(
 {
  role: "roleChercheur",
  privileges: [
   {resource: {db: "Ehealth", collection: "Normal"}, actions: [ "find" ] },
   {resource: {db: "Ehealth", collection: "Alerte"}, actions: [ "find" ] }
  ],
  roles: [{ role: "read", db: "Ehealth"}]
 }
);
```
